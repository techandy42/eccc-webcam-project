```python
metar_path = "/space/hall5/sitestore/eccc/mrd/rpnarmp/snow000/NavCan_WxCams/METAR"
provincial_code = ['AB', 'BC', 'MB', 'NF', 'NS', 'NT', 'NU', 'ON', 'QC', 'YT']
provincial_report = [code + '_ASOS.csv' for code in provincial_code]
```

```python
import pandas as pd
import os

reports = {}

for report, code in zip(provincial_report, provincial_code):
    reports[code] = pd.read_csv(os.path.join(metar_path, report))
```

```python
unique_codes = []
for province_code in reports.keys():
    unique_codes += list(reports[province_code]['wxcodes'].unique())
unique_codes = list(set(unique_codes))
len(unique_codes)
```

```python
unknown_codes = []

for unique_code in unique_codes:
    if not (('M' in unique_code) or ('TS' in unique_code) or ('SN' in unique_code) or ('RA' in unique_code and 'FZ' not in unique_code) or ('GR' in unique_code and 'SGRA' not in unique_code) or ('GS' in unique_code) or ('IP' in unique_code) or ('DZ' in unique_code and 'FZ' not in unique_code) or ('FZRA' in unique_code or 'FZDZ' in unique_code)):
        unknown_codes.append(unique_code)
        
print(unknown_codes)
```

```python
def check_for_M(report):
    report['M'] = report['wxcodes'] == 'M'
    return report

def check_for_TS(report):
    report['TS'] = report['wxcodes'].str.contains('TS')
    return report

def check_for_SN(report):
    report['SN'] = report['wxcodes'].str.contains('SN')
    return report

def check_for_RA(report):
    report['RA'] = report['wxcodes'].str.contains('RA') & ~report['wxcodes'].str.contains('FZ')
    return report

def check_for_Hail(report):
    report['Hail'] = report['wxcodes'].str.contains('GR') & ~report['wxcodes'].str.contains('SGRA')
    return report

def check_for_GS(report):
    report['GS'] = report['wxcodes'].str.contains('GS')
    return report

def check_for_IP(report):
    report['IP'] = report['wxcodes'].str.contains('IP')
    return report

def check_for_DZ(report):
    report['DZ'] = report['wxcodes'].str.contains('DZ') & ~report['wxcodes'].str.contains('FZ')
    return report

def check_for_FZRA_FZDZ(report):
    report['FZRA/FZDZ'] = report['wxcodes'].str.contains('FZRA') & report['wxcodes'].str.contains('FZDZ')
    return report
```

```python
def extract_wxcodes_text_from_report(report):
    check_for_M(report)
    check_for_TS(report)
    check_for_SN(report)
    check_for_RA(report)
    check_for_Hail(report)
    check_for_GS(report)
    check_for_IP(report)
    check_for_DZ(report)
    check_for_FZRA_FZDZ(report)
```

```python
# Categories:
# 1. Missing/No Precipitation
# 2. Snow Only
# 3. Rain Only
# 4. Rain and Snow, But No Ice
# 5. Drizzle Only
# 6. Ice
# 7. Freezing Precipitation

def join_codes(report):
    report['precipitation_category'] = None
    report.loc[report['M'] == True, 'precipitation_category'] = "MISSING"
    report.loc[report['SN'] == True, 'precipitation_category'] = "SNOW_ONLY"
    report.loc[report['RA'] == True, 'precipitation_category'] = "RAIN_ONLY"
    report.loc[(report['RA'] == True) & (report['SN'] == True), 'precipitation_category'] = "RAIN_AND_SNOW"
    report.loc[report['DZ'] == True, 'precipitation_category'] = "DRIZZLE_ONLY"
    report.loc[(report['Hail'] == True) | (report['GS'] == True) | (report['IP'] == True), 'precipitation_category'] = "ICE"
    report.loc[report['FZRA/FZDZ'] == True, 'precipitation_category'] = "FREEZING_PRECIP"
```

```python
for province_code in reports:
    extract_wxcodes_text_from_report(reports[province_code])
    join_codes(reports[province_code])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def graph_categories(df, filename='precipitation_categories_distribution.png'):
    # Count the occurrences of each unique value excluding NaN values
    value_counts = df['precipitation_category'].dropna().value_counts()
    
    print(value_counts)
    
    # Plotting
    value_counts.plot(kind='bar')
    plt.title('Occurrences of Unique Values in column_name')
    plt.xlabel('Unique Values')
    plt.ylabel('Occurrences')
    plt.xticks(rotation=45)  # Rotate labels to make them readable
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()  # Close the plot to prevent it from displaying in the notebook
```

```python
graph_categories(pd.concat([reports[province_code] for province_code in reports], ignore_index=True))
```

```python
import matplotlib.pyplot as plt
import itertools

def graph_precipitation_type_distribution(df, filename='precipitation_distribution.png'):
    # Calculate total occurrences of any precipitation type
    total = df[['M', 'TS', 'SN', 'RA', 'Hail', 'GS', 'IP', 'DZ', 'FZRA/FZDZ']].any(axis=1).sum()
    
    non_m_types = ['TS', 'SN', 'RA', 'Hail', 'GS', 'IP', 'DZ', 'FZRA/FZDZ']
    
    non_m_types_combinations = []
    
    non_m_types_combinations.append(['M'])
    
    for r in range(1, len(non_m_types) + 1):
        for combination in itertools.combinations(non_m_types, r):
            non_m_types_combinations.append(list(combination))
    
    counts = {}
    
    for combination in non_m_types_combinations:
        count = df[combination].all(axis=1).sum()
        if count > 0:
            counts['_'.join(combination)] = count
    
    names = list(counts.keys())
    values = list(counts.values())
    
    # Plotting
    plt.figure(figsize=(10, 8))
    bars = plt.bar(names, values)
    
    plt.xlabel('Precipitation Types')
    plt.ylabel('Count')
    plt.title('Distribution of Precipitation Types')
    plt.xticks(rotation=45, ha="right")
    
    # Adding the total counts above the bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height}', ha='center', va='bottom')
    
    plt.tight_layout()
    
    # Save the figure
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()  # Close the plot to prevent it from displaying in the notebook
```

```python
graph_precipitation_type_distribution(pd.concat([reports[province_code] for province_code in reports], ignore_index=True))
```
